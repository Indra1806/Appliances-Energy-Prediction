## Feature Correlation Analysis

### Subtask:
Calculate the correlation matrix between the 'Appliances' target variable and all other numerical features. Visualize this correlation using a heatmap to identify the most influential predictors, and ensure legends are properly displayed.


**Reasoning**:
To analyze the correlation between 'Appliances' and other numerical features, I will first select all numerical columns, calculate their correlation matrix, and then visualize the correlations with 'Appliances' using a heatmap as instructed, ensuring proper labeling and display.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select only the numerical columns from the DataFrame `df`
numerical_df = df.select_dtypes(include=['number'])

# 2. Calculate the Pearson correlation matrix for these numerical columns
correlation_matrix = numerical_df.corr()

# 3. Extract the correlations specifically with the 'Appliances' target variable
appliances_correlations = correlation_matrix[['Appliances']].sort_values(by='Appliances', ascending=False)

# 4. Create a heatmap of these correlations
plt.figure(figsize=(10, 12))
sns.heatmap(appliances_correlations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, linecolor='black')

# 5. Add a title to the heatmap
plt.title('Correlation Matrix with Appliances Energy Consumption', fontsize=16)
plt.ylabel('Features', fontsize=12)

# 6. Display the plot
plt.show()
```
